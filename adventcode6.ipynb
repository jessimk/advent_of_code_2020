```python
customs = open('input_data/adventcode6.txt', 'r')

line_ct = 0
idnum = 0
result = pd.DataFrame()

processed_forms = []

a_single_form = {'group_id':'',
                'responses':[],
                'yes_count':''}

for line in customs: #each line reps 1 preson
    personal_response = list(line)
    personal_response.remove('\n')
    
    if personal_response != []:
        a_single_form['group_id'] = idnum
        a_single_form['responses']=a_single_form['responses']+personal_response

        line_ct +=1 #increment line count
        
    else:

        a_single_form['yes_count'] = len(set(a_single_form['responses']))
        
        processed_forms.append(a_single_form)
        
        a_single_form = {'group_id':'',
                        'responses':[],
                        'yes_count':''}
        
        idnum+=1 #increment group id
        line_ct +=1 #increment line count
        
if line_ct == 2199:
    a_single_form['yes_count'] = len(set(a_single_form['responses']))
        
    processed_forms.append(a_single_form)
        
result = pd.DataFrame(processed_forms)
    
customs.close()
```

## Puzzle 2

```python
customs = open('input_data/adventcode6.txt', 'r')

line_ct = 0
idnum = 0
result = pd.DataFrame()

processed_forms = []

a_single_form = {'group_id':'',
                 'person_id':''}

for line in customs: #each line reps 1 preson
    personal_response = list(line)
    personal_response.remove('\n')
    #print(personal_response)
    if personal_response != []: #if response line, record
        
        a_single_form['group_id'] = idnum
        a_single_form['person_id'] = line_ct
        
        for item in personal_response:
            a_single_form[item] = personal_response.count(item)
        
        #print(a_single_form)
        processed_forms.append(a_single_form)
        
        a_single_form = {'group_id':'',
                         'person_id':''}
        
        line_ct +=1 #increment line count
        
    else: #if empty line, clear, increment group and line
        #print(processed_forms)
        a_single_form = {'group_id':'',
                         'person_id':''}
        
        idnum+=1 #increment group id
        line_ct +=1 #increment line count

if line_ct == 2199:
    a_single_form['group_id'] = idnum
    a_single_form['person_id'] = line_ct

    for item in personal_response:
        a_single_form[item] = personal_response.count(item)

    #print(a_single_form)
    processed_forms.append(a_single_form)
  
        
result = pd.DataFrame(processed_forms)
    
customs.close()
```

```python
group_responses = result.groupby('group_id').count()
group_responses.head()
```

```python
group_tallies = []
group_summary = {'person_count':'',
             'group_yes_count':''}  

for each_group in np.arange(0,len(group_responses),1):
    person_count = group_responses.loc[each_group].to_list()[0]
    
    all_yes_count = group_responses.loc[each_group].to_list()[1:].count(
        group_responses.loc[each_group].to_list()[0])
    
    group_summary['person_count'] = person_count
    group_summary['group_yes_count'] = all_yes_count
    
    group_tallies.append(group_summary)
    
    group_summary = {'person_count':'',
             'group_yes_count':''} 
    
group_tally_df = pd.DataFrame(group_tallies)
    
```
