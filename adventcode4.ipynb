```python
passport_batch = open('input_data/adventcode4.txt', 'r')

line_ct = 0
idnum = 0

processed_passports = []

a_single_passport = {
             'byr' :"",
            'iyr' :"",
            'eyr' :"",
            'hgt' :"",
            'hcl' :"",
            'ecl' :"",
            'pid' :"",
            'cid' :""}

for line in passport_batch:
    
    line_items = line.split()
    
    if line_items == []:
        processed_passports.append(a_single_passport)
        
        a_single_passport = {
            'byr' :"",
            'iyr' :"",
            'eyr' :"",
            'hgt' :"",
            'hcl' :"",
            'ecl' :"",
            'pid' :"",
            'cid' :""}
        
        idnum+=1
        line_ct +=1
        
    else:
        
        for item in line_items:
            k, v = item.split(':',1)
            a_single_passport[k] = v
            
        line_ct +=1
            
if line_ct == 1096:
    processed_passports.append(a_single_passport)
        
result = pd.DataFrame(processed_passports)
    
passport_batch.close()
```
